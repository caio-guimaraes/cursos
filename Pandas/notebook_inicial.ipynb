### Diferentes tipos de dados

O Pandas fornece diversas funções para importar e exportar dados em diferentes formatos. As principais funções do Pandas para importar dados são as seguintes:

```read_csv()```: Essa função é usada para ler arquivos CSV (Comma Separated Values), que são arquivos de texto que contêm dados separados por vírgulas. É possível passar diversos parâmetros para personalizar a leitura do csv, como delimitador, cabeçalho, tipo de encoding, entre outros.

```read_excel()```: Essa função é usada para ler arquivos do Excel (.xls ou .xlsx) e criar um DataFrame a partir dos dados.

```read_json()```: Essa função é usada para ler arquivos JSON (JavaScript Object Notation), que são arquivos de texto que contêm dados em formato de objeto JavaScript.

```read_html()```: Essa função é usada para ler tabelas HTML, que são estruturas de dados organizadas em formato de tabela em uma página da web.

```read_sql()```: Essa função é usada para ler dados de um banco de dados relacional, como o MySQL, PostgreSQL e SQL Server. O Pandas é capaz de importar dados de diferentes formas, permitindo ajustar parâmetros como a consulta, o nome da tabela e o tipo de dados das colunas.

### Método mean()
O Pandas possui um método chamado ```mean()``` que nos permite calcular a média de um conjunto de dados. Vamos começar calculando a média dos valores da coluna "Valor". Passaremos dados, selecionaremos a coluna 'Valor', e adicionaremos ```.mean()```

````python
dados['Valor'].mean()
````

A média de preços de aluguéis é a média de valores agrupados por tipo: a média de valores de aluguéis de apartamentos, casas e assim sucessivamente. Para isso, temos que agrupar nossos dados conforme o tipo de imóvel. O método que iremos utilizar é o ```groupby()```, que realiza esse agrupamento por alguma coluna específica do dataframe

````python
dados.groupby('Tipo')
````

O método ```groupby()``` funciona em conjunto com outras funções do Pandas, como a ```sum``` de soma ou a ```mean``` que é a média. Com essas funções, podemos fazer agregações conforme o agrupamento que realizamos.

Aplicaremos a função de média, afinal, é isso que queremos calcular quando o agrupamento for feito

````python
dados.groupby('Tipo').mean()
````

Nesse ponto, é importante nos atentarmos a um detalhe. Analisando rapidamente as colunas do nosso dataframe, notaremos que temos tanto colunas numéricas quanto colunas de texto. No código acima, fizemos um agrupamento e aplicamos um método de agregação, o ```mean()```.

Se você estiver usando a versão 1 da biblioteca Pandas, automaticamente esse método de agregação será aplicado unicamente nas colunas numéricas. No entanto, se você estiver utilizando da versão 2 em diante, esse comportamento não será o mesmo. Nesse caso, ao fazer o agrupamento e aplicar o método de agregação, ele será aplicado a todas as colunas, incluindo as de texto. Isso pode gerar um erro.

Para evitar que isso aconteça, podemos especificar um parâmetro no método ```mean()``` (e nos outros métodos de agregação também): ```numeric_only=True```. Com isso, especificamos que a média será calculada apenas nas colunas numéricas e não teremos nenhum erro. Então:

````python
dados.groupby('Tipo').mean(numeric_only=True)
````

Ao executar essa célula, estamos agrupando nosso dataframe dados conforme a coluna Tipo e calculando a média de valores conforme esse agrupamento.

O retorno é um dataframe onde os índices são os tipos de imóveis presentes na nossa base de dados, porque agrupamos nossos dados a partir dessa coluna. As outras colunas possuem a média por tipo de imóvel de valores referentes à coluna em questão do dataframe original.

Por exemplo, os imóveis do tipo "Apartamento" possuem uma média de 2.48 quartos, uma média de 1.23 vagas de garagem e assim sucessivamente. A coluna que realmente importa para nós é a de "Valor", pois é ela que apresenta o valor médio do aluguel por tipo de imóvel.

Os imóveis do tipo "Apartamento" possuem uma média de preço de aluguel de R$4744. Também podemos visualizar as médias de preço dos outros tipos de imóveis do dataframe.

### Melhorando a visualização do dataframe
Podemos selecionar apenas a coluna de valor no dataframe, já que é apenas ela que nos interessa no momento. Vamos copiar o código em que realizamos o agrupamento e calculamos a média. Podemos cselecionar a coluna de valor abrindo colchetes antes do método ```mean()``` e passando seu nome: ```['Valor']```.

Com isso, estamos selecionando uma coluna numérica específica antes da aplicação do método ```mean()```. Isso significa que esse método não será mais aplicado a todas as colunas do dataframe, mas apenas na coluna selecionada. Então, não precisamos mais utilizar o parâmetro ```numeric_only=True``` no método. Podemos apagá-lo, ainda que deixá-lo não vá gerar nenhum erro.

````python
dados.groupby('Tipo')['Valor'].mean()
````

O retorno desse comando apresenta apenas os valores da coluna "Valor" no formato de series. Como são muitas informações, a visualização não ficou tão boa.

Podemos transformar essa visualização em um dataframe. Para isso, basta copiar o código acima, colar na próxima célula e passar colchetes duplos no momento de seleção da coluna valor: ```[['Valor']]```.

A visualização em dataframe ficou bem melhor! Além disso, os dados da coluna "Valor" estão ordenados da menor média de preços de aluguéis para a maior.

Como temos muitos dados, a visualização ainda não é a ideal. Ficar analisando cada um dos números conforme o tipo é trabalhoso. Para facilitar, podemos plotar um gráfico utilizando esse dataframe

#### Aprendendo mais sobre o método GroupBy

O método ```groupby``` do Pandas permite agrupar e resumir dados de um DataFrame, com base em um ou mais critérios. Esses critérios podem ser variáveis numéricas ou categóricas, como colunas ou índices do DataFrame.

A ideia por trás do ```groupby``` é dividir os dados em grupos com base nos critérios selecionados e, em seguida, aplicar uma operação a esses grupos. Essa operação pode ser uma função de agregação, como soma, média, contagem, desvio padrão, entre outras, ou mesmo uma operação personalizada definida pela pessoa usuária.

Esse método possui diversos parâmetros, alguns deles são:

* ```by```: esse é o parâmetro mais comum e é usado para especificar a coluna ou colunas que queremos agrupar. Como argumento dele, podemos passar o nome de uma coluna ou uma lista de nomes de colunas;
* ```axis```: utilizamos esse parâmetro para especificar o eixo ao longo do qual queremos agrupar. O valor padrão dele é 0, o que significa que as linhas serão agrupadas. Se quisermos agrupar as colunas, devemos definir esse parâmetro como 1;
* ```sort```: esse parâmetro é um booleano (True ou False) que indica se os grupos devem ser ordenados pelo valor da coluna de agrupamento. O valor padrão é True;

* ```dropna```: utilizamos esse parâmetro para controlar se os valores ausentes (NaN) serão excluídos ou não durante o processo de agrupamento. O valor padrão é True.

````python
import pandas as pd

df = pd.DataFrame({
   'Animal': ['Cachorro', 'Gato', 'Elefante', 'Cachorro', 'Gato', 'Elefante'],
   'Cor': ['Preto', 'Branco', 'Cinza', 'Marrom', 'Preto', 'Marrom'],
   'Quantidade': [2, 3, 1, 4, 2, 2]
})
df
````

Aqui nós temos um DataFrame com dados referentes a alguns animais. Vamos utilizar o ```groupby``` para agrupar os dados pelo tipo de animal e realizarmos a soma da coluna "Quantidade" para conferirmos o total de cada tipo de animal:

````python
df.groupby('Animal').sum(numeric_only=True)
````

Podemos também usar o método ```groupby``` para agrupar os dados por múltiplas colunas. Por exemplo, vamos agrupar pelo tipo do animal e também pela cor e realizar a soma da coluna "Quantidade":

````python
df.groupby(['Animal', 'Cor'])[['Quantidade']].sum()
````

Assim, podemos observar que o ```groupby``` é um método muito útil para situações em que precisamos dividir os dados em grupos para analisar o comportamento desses grupos. Utilizando esse método, nós podemos responder perguntas específicas sobre nossos conjuntos de dados

## Lidando com dados nulos

Como nossos dados serão usados pelo time de ML para criação de um modelo de precificação de imóveis, é importante tratarmos os dados nulos, uma vez que dados nulos não podem ser utilizados no treinamento de modelos de ML.

Para descobrir se a nossa base de dados possui valores nulos podemos utilizar o método ```df.isnull()```. O retorno será uma tabela , onde o valor false de cada posição representa que o dado não é nulo, e true representa que aquela posição possui valor nulo

````python
df.isnull()
````

Pelo fato do nosso dataframe ser enorme e possuir várias linhas, essa visualização não é a ideal para identificarmos quais colunas possuem dados nulos e quais não possuem.

Para que possamos visualizar isso de forma mais explícita, podemos usar a função sum. Esta soma todos os true e false e nos retorna a quantidade de dados nulos existentes em cada coluna. Os verdadeiros correspondem ao valor 1, e os falsos ao valor 0. Assim, quando fizermos a soma de todos esses valores, vamos ter o total de valores nulos por coluna

````python
df.isnull().sum()
````

Temos o nome de cada coluna à esquerda, e à direita a quantidade de dados nulos existentes em cada uma delas. Notamos que somente as colunas valor, condomínio e IPTU possuem valores nulos. Com isso identificado, precisamos tratar esses dados nulos

Como as colunas que possuem valores nulos são numéricas, vamos substituir os valores nulos pelo valor zero. Vale ressaltar que nem sempre essa é a melhor abordagem e que esse tratamento de valores nulos sempre irá depender do projeto que estamos trabalhando.

Para aplicar essa substituição, usamos um método chamado de ```fillna()```

````python
df.fillna(0)
````

````python
df = df.fillna(0)
````

````python
df.isnull().sum()
````

Podemos conferir que nenhuma das colunas possuem valores nulos

Tratar dados nulos em uma base de dados é muito importante porque esses dados faltantes podem interferir diretamente na qualidade das análises e nos resultados obtidos. Quando existem dados nulos em um conjunto de dados, podem ocorrer problemas como viés, redução da acurácia e imprecisão nas informações obtidas.

Além disso, algumas análises e modelos estatísticos não são capazes de lidar com a presença de dados faltantes. Por isso, o tratamento dos dados nulos é fundamental para garantir a confiabilidade e a precisão das análises e decisões tomadas a partir desses dados.

Existem diversas formas de tratar dados nulos com o Pandas. Algumas das principais formas são:
* Remover os dados nulos: É possível remover as linhas ou colunas que possuem valores nulos utilizando o método ```dropna()``` . Esse método remove todas as linhas ou colunas que possuem pelo menos um valor nulo.
* Preencher os dados nulos: Utilizando o método ```fillna()```, podemos preencher os valores nulos com um valor específico. Além disso, também é possível utilizar argumentos específicos do método ```fillna()``` como o ```method=”ffill”``` ou ```method=”bfill”``` para preencher os valores nulos com o valor anterior ou posterior, respectivamente.
* Interpolar os dados nulos: É possível utilizar o método ```interpolate()``` para preencher os valores nulos com valores interpolados, ou seja, valores calculados a partir dos valores vizinhos.

É importante escolher a melhor forma de tratar os dados nulos de acordo com a situação específica e o objetivo da análise de dados.

# Manipulando os dados

O time de desenvolvedores do site da empresa solicitou a criação de duas novas colunas numéricas na nossa base de dados. No site dos imóveis, eles gostariam de apresentar informações mais sumarizadas em relação a valores. Para isso, essas colunas serão importantes:

* valor_por_mes: essa coluna deve conter os gastos mensais de cada imóvel, incluindo aluguel e condomínio;
* valor_por_ano: essa coluna deve conter os gastos anuais por imóvel, ou seja, IPTU mais 12 meses de aluguel e condomínio.

## Criando colunas numéricas

Começaremos importando novamente nossa base de dados. Isso porque, para o site da imobiliária, as informações de todos os imóveis são importantes. Mas lembre-se que, anteriormente, trabalhamos apenas com apartamentos, então não faz sentido continuar utilizando o mesmo dataframe.

````python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
````

Vamos começar criando a coluna numérica ```valor_por_mes```

````python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
````

Perceba que, ao final do dataframe, foi adicionada a coluna "Valor_por_mes" com a soma realizada. Criamos a primeira coluna numérica! Contudo, podemos notar que, a operação não foi efetuada para as linhas que possuem valores ```NaN```. Precisamos realizar o tratamento dessas informações e repetir o processo anterior:

````python
dados.fillna(0, inplace = True)
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
````

Agora, criaremos "Valor_por_ano" seguindo um procedimento semelhante. Para esta coluna, devemos considerar os gatos mensais multiplicados por 12 (meses do ano) somado ao valor do IPTU

````python
dados['Valor_por_ano'] = dados['IPTU'] + (12 * dados['Valor_por_mes'])
dados.head()
````

## Criando colunas categóricas

A primeira coluna será chamada de "Descrição" e deve conter uma sumarização das principais informações de cada um dos imóveis da nossa base de dados. Entre elas é necessário constar o tipo de imóvel, o bairro, a quantidade de quartos e de vagas de garagem. A segunda coluna se chamará "Possui_suite" e será binária, onde devemos informar se o imóvel tem suíte ou não.

````python
dados['Descricao'] = dados['Tipo'] + dados['Bairro']
dados.head()
````

Note que as informações ficaram grudadas, sem nenhum espaço, então podemos personalizar acrescentando o termo "em", com espaço antes e depois, entre o tipo e o bairro:

````python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro']
dados.head()
````

Vamos incluir as outras informações. Iremos utilizar a barra invertida ```\``` para quebrar linha e facilitar a visualização do comando

````python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'] + ' quarto(s) ' + \
                                        ' e ' + dados['Vagas'] + ' vaga(s) de garagem.'
````

Ao executar, note que temos um TypeError, porque não é possível concatenar strings com números inteiros, que é o caso das colunas numéricas de quartos e vagas. Sendo assim, é necessário converter essas informações para string usando o método .```asType()``` e informando para qual tipo será feita a conversão.

````python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
````

````python
dados.head()
````

Vamos criar agora a coluna "Possui_suite"

Para criá-la, precisamos analisar a coluna já existente que possui a quantidade de suítes em cada um dos imóveis. Sendo assim, teremos que analisar cada linha dessa coluna e colocar "Sim" para as que tiverem valor maior que zero, e "Não" para o caso contrário.

Usaremos o método ```.apply()```, utilizado para aplicar uma determinada função em um conjunto específico de dados. No nosso caso, esse conjunto é a coluna "Suites", e a função, é do tipo lambda - funções personalizadas que podem ser escritas em uma única linha

````python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
````

Nesta função, ```x``` equivale a cada linha da coluna de suítes. Então, fizemos uma condicional ```if``` em que se o valor de ```x```, ou seja, o registro da linha, for maior que zero, deve ser colocado "Sim", caso contrário (```else```), coloca-se "Não"

````python
dados.head()
````

Agora, vamos salvar o dataframe em arquivo CSV chamado "dados_completos_dev"

````python
dados.to_csv('dados_completos_dev.csv', index=False, sep=';')
````

#### Outras formas de criar colunas

Existem diversas maneiras de criar colunas em um DataFrame no Pandas. Alguns exemplos são:

1 - Atribuição direta de valores a uma nova coluna:

````python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = [7, 8, 9]
df
````

2 - Criação de uma coluna a partir de operações entre outras colunas:

````python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = df['A'] + df['B']
df
````

3 - Utilização do método ```assign()``` para criar uma nova coluna:

````python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df = df.assign(C=[7, 8, 9])
df
````

4 - Utilização do método ```apply()``` para aplicar uma função a uma coluna existente e criar uma nova coluna:

````python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = df['A'].apply(lambda x: x * 2)
df
````
