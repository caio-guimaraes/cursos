# Para saber mais: formato wide e long

Os dados em uma tabela podem estar organizados em formatos diferentes de acordo com o modo em que foram preenchidos ou da necessidade da anlise que ser feita. Os termos wide e long so usados para descrever a organizao dos dados em um banco de dados relacional.

No formato **wide**, tambm conhecido como tabelas de piv, as tabelas tm uma nica linha de cabealho contendo diversas colunas, geralmente mais colunas do que linhas, e so teis para armazenar informaes que possuem muitos atributos, como questionrios e pesquisas. As colunas podem conter valores ausentes para as clulas em que os dados no se aplicam.

No formato **long**, tambm conhecido como formato normalizado ou estreito, as tabelas armazenam cada observao em uma linha separada. Ao invs de ter uma nica linha de cabealho contendo todas as colunas, as tabelas no formato long geralmente tm duas colunas principais de informao, uma para os nomes das variveis e outra para os valores das variveis. So teis para armazenar informaes com muitas observaes, como sries temporais ou dados de experimentos cientficos.

Na biblioteca pandas,  possvel fazer uma mudana do formato wide para o formato long com o uso do mtodo melt(). A sintaxe para utilizar o mtodo melt()  a seguinte:

```
dataframe.melt(id_vars = [...], value_vars = [...], var_name = '..', value_name = '..')
```

O parmetro ```id_vars```  utilizado para identificar os nomes das colunas que sero mantidas no momento da transformao. J o parmetro ```value_vars```  utilizado para identificar as colunas que sero transformadas nas colunas principais do formato long, uma contendo as categorias com os nomes das colunas e outra contendo os valores. O parmetro ```var_name```  usado para nomear a coluna contendo as categorias e o parmetro ```value_name``` para nomear a coluna com os valores numricos.
