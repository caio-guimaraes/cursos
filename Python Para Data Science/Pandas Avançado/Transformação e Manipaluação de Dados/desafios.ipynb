- **Projeto desafio 1: vendas online**

````python
colunas = list(vendas_clientes.columns)
colunas

vendas_clientes = vendas_clientes.explode(colunas[1:])
vendas_clientes.reset_index(drop=True,inplace=True)
vendas_clientes
````

````python
vendas_clientes['Valor da compra']
````

````python
colunas = list(locacao_imoveis.columns)
colunas

locacao_imoveis = locacao_imoveis.explode(colunas[1:])
locacao_imoveis.reset_index(drop=True,inplace=True)
locacao_imoveis
````

````python
import numpy as np
locacao_imoveis['valor_aluguel'] = locacao_imoveis['valor_aluguel'].apply(lambda x: x.replace('$ ', '').replace(' reais', '').replace(',','.').strip())
locacao_imoveis['valor_aluguel'] = locacao_imoveis['valor_aluguel'].astype(np.float64)
locacao_imoveis.info()
````
