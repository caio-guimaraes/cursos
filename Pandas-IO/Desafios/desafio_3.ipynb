# Desafio: normalizando o JSON obtido na API

Chegou a hora de testar os conhecimentos desenvolvidos durante a aula. O DataFrame mostrado abaixo foi gerado aps a obteno de dados da API [Fruitvice](https://fruityvice.com/):



Este DataFrame possui 6 colunas: genus (gnero), name (nome), id, family (famlia), order (ordem) e nutritions (nutries). Note que a coluna nutritions est com todas as informaes nutricionais aninhadas. Portanto, os dados precisam ser normalizados.

O desafio agora  normalizar esse DataFrame.

```python
import requests
import json

dados_frutas = requests.get('https://fruityvice.com/api/fruit/all')
resultado = json.loads(dados_frutas.text)
```

```python
import pandas as pd
dados_frutas_normalizado = pd.json_normalize(resultado)
```
