## 2 -Plotly express x Graph objects
Aqui, vamos explorar duas das principais bibliotecas do Plotly: a [Plotly Express](https://plotly.com/python/plotly-express/) e a [Graph Objects](https://plotly.com/python/graph-objects/).

Sabemos que a Plotly possui uma incrvel capacidade de criar grficos impactantes e interativos, que vo desde simples disperses at visualizaes complexas de dados multidimensionais. Dito isto, temos duas abordagens distintas para a criao de grficos com o Plotly que so a partir da utilizao da biblioteca **Plotly Express** ou da **Plotly Graph Objects**.

### Plotly Express
Com uma abordagem de alto nvel, a **Plotly Express** (usualmente importada como px) oferece simplicidade e rapidez na criao de grficos interativos, sendo ideal para anlise exploratria de dados. Entre seus pontos positivos, podemos destacar:
- Interface de alto nvel para criao rpida e fcil de grficos interativos.
- Funes simples e diretas para criar uma variedade de grficos, como grficos de disperso, barras, linhas, bolhas, entre outros.
- Requer menos cdigo e  mais amigvel para iniciantes.
- Criao rpida de visualizaes interativas sem a necessidade de ajustes detalhados.

Todos esses pontos apontam a Plotly Express como a ideal para prototipagem rpida e anlise exploratria de dados. Mas, temos muitas vantagens tambm em utilizar a Graph Objects como podemos conferir a seguir.

## Graph Objects
A **Plotly Graph Objects** (usualmente importada como *go*) oferece um nvel mais avanado de controle, possibilitando a personalizao detalhada e a criao de grficos complexos sob medida para os requisitos especficos do projeto. Entre seus pontos positivos, podemos destacar:

Interface de baixo nvel que oferece controle detalhado sobre cada elemento do grfico.
Requer mais cdigo e conhecimento mais avanado para criar grficos personalizados.
Construo de grficos complexos com total controle sobre layouts, anotaes e estilos.
Adequada para personalizao avanada e criao de grficos que no so facilmente alcanados com o Plotly Express.
Todos esses pontos apontam a Plotly Graph Objects como ideal para visualizaes personalizadas, relatrios especficos e quando  necessrio ter controle total sobre o design dos grficos.

### Como escolher entre as duas bibliotecas
Ao entender as diferenas entre essas duas bibliotecas, as pessoas cientistas de dados e desenvolvedoras podem tomar decises sobre qual abordagem utilizar para criar visualizaes de dados impactantes e eficazes. Assim, podemos elencar trs situaes que podem te dar um auxlio sobre qual utilizar:

Se voc precisa criar rapidamente grficos interativos com uma sintaxe simples, escolha Plotly Express.
Se voc busca criar grficos altamente personalizados com controle detalhado ou precisa construir visualizaes complexas, use Plotly Graph Objects.
Em alguns casos,  possvel combinar ambas as bibliotecas para aproveitar o melhor de cada uma, comeando com Plotly Express para uma prototipagem rpida e, em seguida, migrando para Plotly Graph Objects para ajustes finos e personalizaes.
Observe um exemplo da criao de um grfico de colunas utilizando essas bibliotecas separadamente.

Exemplo de um grfico de colunas para ambas bibliotecas
Vamos representar, em um grfico de colunas, as vendas por unidade dos sorvetes da Serenatto - Caf & Bistr, a franquia de restaurantes mencionada no exerccio anterior, utilizando as duas bibliotecas que estamos estudando.

**Utilizando a Plotly Express:**

```python
import plotly.express as px

# Dados de exemplo
data = {'sabor': ['Chocolate', 'Baunilha', 'Morango', 'Limo', 'Caj', 'Uva'], 
        'qtd': [180, 170, 100, 90, 60, 40]}

# Criando um grfico de colunas usando Plotly Express
fig = px.bar(data, x='sabor', y='qtd', title='Vendas em unidades do sorvete da Serenatto - Caf & Bistr',
             labels={'sabor': 'Sabor', 'qtd': 'Quantidade'}, text_auto=True,
             hover_data=['sabor','qtd'])

# Mostrando o grfico
fig.show()
```

**Utilizando a Plotly Graph Objects:**

```python
import plotly.graph_objects as go

# Dados de exemplo
data = {'sabor': ['Chocolate', 'Baunilha', 'Morango', 'Limo', 'Caj', 'Uva'], 
        'qtd': [180, 170, 100, 90, 60, 40]}

# Criando um grfico de colunas usando Plotly Graph Objects
fig = go.Figure()

fig.add_trace(go.Bar(x=data['sabor'], y=data['qtd'], text=data['qtd'], textposition='auto'))

fig.update_layout(title='Vendas em unidades do sorvete da Serenatto - Caf & Bistr',
                  xaxis_title='Sabor', yaxis_title='Quantidade')

# Mostrando o grfico
fig.show()
```

Note que ambos os grficos gerados sero semelhantes.

Resumindo, ambas as bibliotecas so boas ferramentas do Plotly para visualizaes interativas. O Express  indicado para grficos rpidos e fceis, enquanto o Graph Objects oferece controle avanado para visualizaes personalizadas. A escolha depende da agilidade versus personalizao desejada. Combin-las  uma estratgia eficaz para criar visualizaes sofisticadas e geis.

## 3 - normalizao de dados
Vamos pontuar uma das razes de explorarmos a normalizao mais voltada para criao de visualizaes de dados. A normalizao de dados tenta resolver o problema da criao de visuais cujos dados possam representar diferenas muito grandes em uma mesma varivel tornando difcil a interpretao em nosso grfico. Alm disso, ela  bastante til quando queremos averiguar a composio dos dados em termos percentuais.

A normalizao dos dados  essencial para garantir que as variveis estejam em uma escala comparvel, facilitando a interpretao dos resultados. Em nosso ltimo vdeo, utilizamos uma expresso similar a `df = df.div(df.sum(axis=1), axis=0)` para dividir cada valor pelo somatrio da linha correspondente, realizando a normalizao das linhas do DataFrame. Isso permite que a pessoa cientista de dados transforme seus dados em uma escala de 0 a 1, que pode ser facilmente transformada em uma escala percentual (0 a 100).

### Por que transformar em escala percentual?
A transformao dos dados para a escala percentual pode ser extremamente til para entender seus dados. Por exemplo, se voc estivesse analisando as vendas de vrios produtos, ter a informao em termos percentuais poderia te dizer rapidamente que produto estava gerando a maior parte das vendas.

Transformar seus dados em escala percentual tambm pode ser importante para a normalizao. Voc pode ter dados onde a maioria das observaes esteja em um pequeno intervalo de valores, mas alguns poucos dados muito grandes ou muito pequenos distorcem a escala de tal forma que a maioria deles  espremida em uma pequena parte da escala. Ao normalizar, voc garante que todos os dados so representados de forma equitativa.

### Explicando a frmula de normalizao
Vamos utilizar a mesma frmula que aprendemos na aula para entender como conseguimos normalizar dados na mesma linha. Vamos resgatar a expresso logo abaixo para explic-la

```python
df = df.div(df.sum(axis=1), axis=0)
```

O mtodo `sum()`  aplicado ao longo das linhas (`axis=1`) para calcular a soma de cada linha do DataFrame, atuando como o nosso denominador da diviso. Enquanto que o mtodo div()  utilizado para dividir cada valor do DataFrame pelo valor correspondente na soma da sua respectiva linha utilizando o eixo das colunas (`axis=0`).

>  importante ter ateno na diviso que  feita por eixo das colunas porque os valores esto sendo varridos na mesma linha pulando coluna a coluna. Por isso que na funo `div()`, passamos o parmetro `axis=0`.

Vamos representar em um conjunto de dados normalizados a composio das vendas dos sorvetes da Serenatto - Caf & Bistr de acordo com as 4 principais filiais. Primeiro vamos criar o DataFrame e depois aplicar a normalizao com a frmula explicada mais acima:

## 4 - Grfico de reas empilhadas 100%
O grfico de reas empilhadas 100%  um tipo de grfico de reas em que os dados so apresentados como porcentagens do total, permitindo comparar a distribuio relativa de vrias categorias ao longo de um eixo. Neste tipo de grfico, a rea total da representao  sempre igual a 100%, o que facilita a visualizao das propores entre as categorias.

Este tipo de grfico  til quando desejamos comparar a contribuio de diferentes categorias para o total em vrias sries de dados. Ele  bastante til quando temos vrias sries com tamanhos variados e queremos entender a distribuio percentual de cada categoria em relao ao total. Alm disso, os grficos de reas empilhadas 100% so adequados para mostrar como a composio das categorias muda ao longo do tempo.

Vamos entender as diferenas entre o grfico de reas empilhadas com valores absolutos e o de reas empilhadas 100% por meio de um exemplo explorando um case do e-commerce Buscante, um buscador online de livros.

Neste caso, voc precisa representar a composio dos acessos dos usurios nos ltimos 5 anos (2018 a 2022) na plataforma, tomando como base o acesso realizado nos trs pases em que a empresa possui clientes. A base de dados que voc recebeu possui as quantidades de acesso dos trs pases (Brasil, Argentina e Colmbia) em cada ano de pesquisa.

Para resolver este problema, vamos criar 2 diferentes grficos: o grfico de reas empilhadas para mostrar a magnitude e variao dos acessos ano a ano e um grfico de reas empilhadas 100% para mostrar a distribuio relativa dos acessos entre os pases.

Primeiro, vamos importar as bibliotecas que vamos utilizar e criar o DataFrame com os dados fornecidos para a construo do 1 grfico:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Acessos em milhes de cada pas em cada ano
acessos = pd.DataFrame({'Ano': [2018, 2019, 2020, 2021, 2022], 'Brasil': [120, 150, 180, 200, 220],
                        'Argentina': [85, 110, 130, 150, 170],  'Colmbia': [50, 70, 90, 110, 130]})

acessos = acessos.set_index("Ano")
acessos
```

Em seguida, vamos aplicar a normalizao em `acessos`, trazendo os nmeros para a escala percentual (0-100) a fim de construir o 2 grfico:

```python
# Normalizando os nmeros de acessos e deixando-os numa escala de 0 a 100
acessos_normalizados = (acessos.div(acessos.sum(axis=1), axis=0) * 100).round(1)
acessos_normalizados
```

Por fim, para construir ambos os grficos lado a lado para realizarmos a comparao, vamos executar o seguinte cdigo:

```python
fig, axs = plt.subplots(1, 2, figsize=(14,7))

fig.suptitle('Composio dos acessos dos usurios da Buscante no Brasil, Argentina e Colmbia de 2018 a 2022', 
             fontsize = 16, ha = 'left', x = 0.03, y = 1.05)

# GRFICO 1 - grfico de reas empilhadas

acessos.plot.area(color = ["#009C3B", "#75AADB", "#FFCD00"], xticks=np.arange(2018, 2023,1), ax = axs[0], alpha= 0.6)
axs[0].set_title('Em milhes de views\n', loc = "left")
axs[0].set_ylabel('')
axs[0].set_xlabel('')
axs[0].xaxis.set_tick_params(labelsize=12)
axs[0].yaxis.set_tick_params(labelsize=12)
axs[0].set_ylim(0.01, 570)     # ajustando os limites do eixo y
axs[0].set_xbound(2018,2023)   # ajustando os limites do eixo x
axs[0].legend().remove()
axs[0].tick_params(axis='both', which='both', length=0)
axs[0].set_frame_on(False)

## Anotando os nomes das regies
pais = acessos.columns
y_pos = [80, 240, 350]
for i in range(len(pais)):
  axs[0].text(2020, y_pos[i], f'{pais[i]}', color = '#231F20', weight = "bold", fontsize = 14, ha="center", transform=axs[0].transData)

#######################################

# GRFICO 2 - grfico de reas empilhadas 100%

acessos_normalizados.plot.area(color = ["#009C3B", "#75AADB", "#FFCD00"], xticks=np.arange(2018, 2023,1), ax = axs[1], alpha= 0.6)
axs[1].set_title('Em porcentagem (%)\n', loc = "left")
axs[1].set_ylabel('')
axs[1].set_xlabel('')
axs[1].xaxis.set_tick_params(labelsize=12)
axs[1].yaxis.set_tick_params(labelsize=12)
axs[1].set_ylim(0.01, 110)     # ajustando os limites do eixo y
axs[1].set_xbound(2018,2023)   # ajustando os limites do eixo x
axs[1].legend().remove()
axs[1].tick_params(axis='both', which='both', length=0)
axs[1].set_frame_on(False)

## Anotando os nomes das regies
pais = acessos_normalizados.columns
y_pos = [20, 60, 90]
for i in range(len(pais)):
  axs[1].text(2020, y_pos[i], f'{pais[i]}', color = '#231F20', weight = "bold", fontsize = 14, ha="center", transform=axs[1].transData)

plt.tight_layout()
plt.show()
```

Note, neste exemplo da Buscante, que possumos diferentes interpretaes para os dois grficos. Em essncia os dados so os mesmos, mas as escalas escolhidas possuem intencionalidades diferentes para apresentao do dado.

O grfico de barras empilhadas foca na magnitude dos dados apontando, por exemplo, o crescimento global dos acessos e tambm as diferenas absolutas entre cada pas. J o grfico de reas empilhadas 100% representa a contribuio de cada pas para o total de acessos. Ou seja, no conseguimos averiguar a evoluo ou no dos acessos totais ano a ano, mas  mais evidente a composio de cada pas para os acessos em cada um dos anos.

## 4 - mapa de calor e matriz de correlao
Na fase inicial de um projeto de anlise de dados  bem comum que faamos uma srie de tentativas para que possamos compreender melhor os padres nos dados e verificar se certos valores contribuem ou no para o problema que enfrentamos.

Dito isto, existe uma srie de visualizaes que podemos utilizar para relacionar dados, sendo duas delas bastante comuns para uma pessoa cientista de dados: os **mapas de calor** (*heatmap*) e as **matrizes de correlao**. Elas ajudam a identificar relaes entre variveis, bem como investigar a fora entre as relaes que podemos chamar de correlao dos dados. Mas o que  cada um deles?

### Mapa de calor
O mapa de calor  uma representao grfica que mostra a distribuio, relao ou variao dos dados em uma matriz ou espao bidimensional, utilizando as cores para destacar valores ou como representao de intensidade na distribuio dos dados. Existem vrios tipos de mapas de calor, partindo dos mapas de calor de correlao, que vamos dar maior destaque aqui, aos mapas de calor de densidade, os geogrficos, entre outros.

O mapa de calor pode ser utilizado para uma variedade de situaes, incluindo:
- Visualizar a relao entre duas ou mais variveis (**mapa de calor de correlao**).
- Visualizar a distribuio de dados (**mapa de calor de intensidade**).
- Apresentar a densidade de ocorrncia de um evento em uma rea especfica (**mapa da calor de densidade**).
- Distribuio geogrfica dos dados (**mapa de calor geogrfico**).

Para representar o relacionamento entre variveis, o mapa de calor utiliza cores para destacar padres e variaes nos dados no formato de tabela, mostrando a correlao entre diferentes combinaes de valores. Os eixos x e y representam as dimenses e a intensidade das cores ou tons representam os valores das clulas.

### Matriz de correlao
A matriz de correlao  uma tabela que mostra a relao estatstica entre mltiplas variveis. Ela exibe os coeficientes de correlao entre os pares de variveis, indicando a fora e direo da relao linear entre elas. Os valores variam entre -1 e 1, nos quais:
- -1 indica uma correlao negativa perfeita;
- 1 indica uma correlao positiva perfeita; e
- 0 indica ausncia de correlao.

A matriz de correlao pode ser utilizada para uma variedade de situaes, incluindo:
- Identificar variveis que esto relacionadas.
- Eliminar variveis que no esto relacionadas.
- Auxiliar na escolha de variveis para modelos de aprendizado de mquina.

### Exemplificando os visuais
Podemos notar que os mapas de calor e as matrizes de correlao podem ser utilizados juntos para uma anlise mais completa dos dados. Os mapas so apontados para visualizar a relao dos dados, enquanto as matrizes de correlao podem ser utilizadas para medir a associao entre as variveis.

Vamos criar um simples exemplo para observar essas visualizaes. Temos abaixo uma pequena tabela representando uma lista de 10 funcionrios e suas respectivas idades, salrios e tempo de servio em uma dada empresa:



Para construir a matriz de correlao, vamos passar os dados do DataFrame que gerou a tabela e chamar o mtodo `corr()` da biblioteca Pandas que cria essa visualizao:

```python
import pandas as pd

df = pd.DataFrame({"Idade": [50, 36, 29, 42, 40, 44, 32, 32, 45, 45],
                   "Salrio": [3130, 4685, 3769, 5391, 8611, 9949, 5433, 8311, 8051, 9420],
                   "Tempo de Servio": [1, 12, 10, 6, 13, 12, 9, 1, 11, 11]})

matriz_correlacao = df.corr()
matriz_correlacao
```

A matriz correlao demonstra a relao entre as variveis. Podemos notar, por exemplo, uma pequena correlao positiva entre o salrio e o tempo de servio. Por fim, vamos representar os dados em um mapa de calor, escrevendo o seguinte cdigo:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Criando o heatmap
fig, ax = plt.subplots(figsize=(8, 4))
sns.set(font_scale=1.1)
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', center=0)
plt.title('Mapa de calor da Correlao entre Idade, Salrio e Tempo de Servio\n')
plt.show()
```

Note que utilizamos o mtodo `heatmap()` do Seaborn para construir o mapa de calor. As clulas so coloridas de acordo com o valor da correlao, facilitando a identificao de padres e relaes entre as variveis.
